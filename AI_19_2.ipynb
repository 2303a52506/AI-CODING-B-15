### Step 2: Write your Java code to a `.java` file

Since Colab cells execute Python, you'll need to write your Java code into a file first. I'll use the `%%writefile` magic command to create `FileHandler.java` with the Java code you provided.

```python
%%writefile FileHandler.java

import java.io.*;
public class FileHandler {
    public static void main(String[] args) {
        try {
            // Reading from input.txt
            BufferedReader reader = new BufferedReader(new FileReader("input.txt"));
            StringBuilder data = new StringBuilder();
            String line;
            while ((line = reader.readLine()) != null) {
                data.append(line).append("\n");
            }
            reader.close();

            // Writing to output.txt in uppercase
            BufferedWriter writer = new BufferedWriter(new FileWriter("output.txt"));
            writer.write(data.toString().toUpperCase());
            writer.close();

            System.out.println("File content successfully written in uppercase.");

        } catch (FileNotFoundException e) {
            System.out.println("File not found.");
        } catch (IOException e) {
            e.printStackTrace();
        }
    }
}
```

 **TASK-2.C++ Code for Sorting Algorithm**

Here is a simple C++ implementation of a Bubble Sort algorithm:

**PROMPT**
"Convert this C++ Bubble Sort program into Python."

```python
%%writefile bubble_sort.cpp

#include <iostream>
#include <vector>

// Function to perform Bubble Sort
void bubbleSort(std::vector<int>& arr, int n) {
    for (int i = 0; i < n - 1; i++) {
        for (int j = 0; j < n - i - 1; j++) {
            if (arr[j] > arr[j + 1]) {
                // Swap arr[j] and arr[j+1]
                int temp = arr[j];
                arr[j] = arr[j + 1];
                arr[j + 1] = temp;
            }
        }
    }
}

// Function to print an array
void printArray(const std::vector<int>& arr, int size) {
    for (int i = 0; i < size; i++) {
        std::cout << arr[i] << " ";
    }
    std::cout << std::endl;
}

int main() {
    std::vector<int> arr = {64, 34, 25, 12, 22, 11, 90};
    int n = arr.size();
    std::cout << "Original array: ";
    printArray(arr, n);

    bubbleSort(arr, n);

    std::cout << "Sorted array: ";
    printArray(arr, n);
    return 0;
}
```

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Test the bubble_sort function
if __name__ == "__main__":
    arr = [64, 34, 25, 12, 22, 11, 90]
    print("Original array:", arr)

    bubble_sort(arr)

    print("Sorted array:", arr)
```

**Task 3 – SQL to Pandas Query**

SQL Query (Original):


**PROMPT**:
"Convert this SQL query with GROUP BY and SUM into equivalent Pandas code."

```python
# Load the SQL extension for IPython magic commands
%load_ext sql
```

```python
%%sql sqlite:///:memory:

CREATE TABLE sales_table (
    department TEXT,
    sales INTEGER
);

INSERT INTO sales_table (department, sales) VALUES
('A', 100),
('B', 200),
('A', 150),
('C', 300),
('B', 100);
```

```python
# Now, execute your original SQL query
%%sql
SELECT department, SUM(sales) as total_sales
FROM sales_table
GROUP BY department;
```

PYTHON PANDAS CODE

```python
import pandas as pd

# Sample dataset
data = {
    "department": ["A", "B", "A", "C", "B"],
    "sales": [100, 200, 150, 300, 100]
}

df = pd.DataFrame(data)

# Pandas equivalent of SQL GROUP BY and SUM
result = df.groupby("department")["sales"].sum().reset_index()
result.rename(columns={"sales": "total_sales"}, inplace=True)
print(result)
```

**Task 5 – Basic Function Translation (Python to C++)**

**PYTHON CODE:**

```python
def check_even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even_odd(5))
print(check_even_odd(8))
print(check_even_odd(0))
```

**PROMPT:**
"Convert this Python even/odd function into C++ code."

**C++ Code:**

```python
%%writefile even_odd_checker.cpp

#include <iostream>
#include <string>

// Function to check if a number is even or odd
std::string checkEvenOdd(int n) {
    if (n % 2 == 0)
        return "Even";
    else
        return "Odd";
}

int main() {
    std::cout << checkEvenOdd(5) << std::endl;
    std::cout << checkEvenOdd(8) << std::endl;
    std::cout << checkEvenOdd(0) << std::endl;
    return 0;
}
```
