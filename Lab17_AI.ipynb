## Task 1: Student Data Cleaning

**PROMPT:**
Generate Python code to clean student dataset by handling missing values,
removing duplicates, standardizing text, and encoding categorical variables.



```python
import pandas as pd

# Initial sample data (replace with your actual data if available)
data = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 1],
    'marks': [80, None, 90, 75, 80],
    'attendance': [85, 90, None, 80, 85],
    'department': ['CSE', 'ECE', None, 'CSE', 'CSE'],
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male']
})

# Apply cleaning operations
data['marks'] = data['marks'].fillna(data['marks'].mean())
data['attendance'] = data['attendance'].fillna(data['attendance'].mean())
data['department'] = data['department'].fillna('unknown')
data.drop_duplicates(subset='student_id', inplace=True)
data['gender']=data['gender'].map({'Male':0,'Female':1})
print(data)
```

## Task 2: Transactions Preprocessing

**PROMPT:**
Preprocess financial dataset by converting dates, creating features,
normalizing amounts, and handling extreme values.


```python
from sklearn.preprocessing import MinMaxScaler
df = pd.DataFrame({
 'date':['2023-01-01','2023-02-01','2023-03-01'],
 'amount':[100,5000,300]})
df['date']=pd.to_datetime(df['date'])
df['month']=df['date'].dt.month
scaler=MinMaxScaler()
df['amount']=scaler.fit_transform(df[['amount']])
print(df)
```

## Task 3: Sensor Data

**PROMPT:**
Clean sensor data by handling missing values, scaling features,
encoding categorical fields, and splitting dataset.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sensor = pd.DataFrame({
 'temp':[30,None,25],
 'humidity':[60,70,None],
 'status':['ok','fail','ok']})
sensor.fillna(sensor.mean(numeric_only=True), inplace=True)
sensor['status']=sensor['status'].map({'ok':1,'fail':0})
scaler=StandardScaler()
sensor[['temp','humidity']]=scaler.fit_transform(sensor[['temp','humidity']])
train,test=train_test_split(sensor,test_size=0.2)
print(train)
```

## Task 4: Traffic Data

**PROMPT:**
Clean traffic dataset by standardizing text, handling missing values,
removing duplicates, normalizing metrics, and generating summary.

```python
traffic=pd.DataFrame({
 'road':['A','A','B'],
 'density':[None,50,70],
 'speed':[40,60,80]})
traffic['density'] = traffic['density'].fillna(traffic['density'].mean())
traffic.drop_duplicates(inplace=True)
print(traffic.describe())
```

## Task 5: Text Processing

**PROMPT:**
Clean social media text by removing duplicates, cleaning text,
tokenizing, removing stopwords, and extracting features.

```python
import re
text=pd.DataFrame({'post':['Hello!!!','Visit http://abc.com',None]})
text.dropna(inplace=True)
text['post']=text['post'].str.lower()
text['post']=text['post'].apply(lambda x: re.sub(r'http\S+|[^a-z ]','',x))
text['word_count']=text['post'].apply(lambda x: len(x.split()))
print(text)
```
